```python
from sklearn.cluster import AgglomerativeClustering
from sklearn import datasets
import pandas as pd
```

```python
iris = datasets.load_iris()
X=iris['data']
Y=iris.target
```

```python
X.shape
```

```python
X
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X[:,0],X[:,1],'ro')
```

# CLustering dengan linkage="ward"

```python
clustering = AgglomerativeClustering(n_clusters=3, linkage="ward", affinity='euclidean')
clustering.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1], c=clustering.labels_, marker="o")
plt.show()
```

# Clustering dengan linkade="average"

```python
clustering = AgglomerativeClustering(n_clusters=3, linkage="average", affinity="euclidean")
clustering.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1], c=clustering.labels_, marker="o")
plt.show()
```

# Clustering dengan linkage="complete"

```python
clustering = AgglomerativeClustering(n_clusters=3, linkage="complete", affinity="euclidean")
clustering.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1], c=clustering.labels_, marker="o")
plt.show()
```

# Clustering dengan linkage="average" dan Affinity="manhattan"

```python
clustering = AgglomerativeClustering(n_clusters=3, linkage="complete", affinity="manhattan")
clustering.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1], c=clustering.labels_, marker="o")
plt.show()
```

# Clustering dengan linkage="complete" dan Affinity="manhattan"

```python
clustering = AgglomerativeClustering(n_clusters=3, linkage="complete", affinity="manhattan")
clustering.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1], c=clustering.labels_, marker="o")
plt.show()
```
